```python
import xml.etree.ElementTree as ET

# Load and parse the .tcx file
tree = ET.parse('./activity_16836104884.tcx')
root = tree.getroot()

# Define the XML namespaces (modify if needed)
namespaces = {'': 'http://www.garmin.com/xmlschemas/TrainingCenterDatabase/v2'}

# Iterate through each Trackpoint and extract data
for activity in root.findall('.//Activity', namespaces):
    sport = activity.get('Sport')
    print(f"Sport: {sport}")
    
    for lap in activity.findall('Lap', namespaces):
        lap_start = lap.get('StartTime')
        print(f"Lap Start Time: {lap_start}")
        
        for track in lap.findall('Track', namespaces):
            for trackpoint in track.findall('Trackpoint', namespaces):
                time = trackpoint.find('Time', namespaces).text
                position = trackpoint.find('Position', namespaces)
                if position is not None:
                    lat = position.find('LatitudeDegrees', namespaces).text
                    lon = position.find('LongitudeDegrees', namespaces).text
                    print(f"Time: {time}, Latitude: {lat}, Longitude: {lon}")
                
                # Extract other data like altitude, heart rate, etc., if present
                altitude = trackpoint.find('AltitudeMeters', namespaces)
                if altitude is not None:
                    print(f"Altitude: {altitude.text} meters")

                heart_rate = trackpoint.find('.//HeartRateBpm/Value', namespaces)
                if heart_rate is not None:
                    print(f"Heart Rate: {heart_rate.text} bpm")
```
